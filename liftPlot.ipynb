```python
import matplotlib.pyplot as plt
import numpy as np
g = 9.82
thiefAcc = np.arange(0., 5., 0.1)

def aditionalMass(colonyMass, thiefAcc):
    return colonyMass*thiefAcc/g

for colonyMass in range(30, 80, 10):
    plt.plot(thiefAcc, aditionalMass(colonyMass, thiefAcc), label='{} kg'.format(colonyMass))

plt.ylabel('additional weight kg')
plt.xlabel('acceleration m/(s^2)')
plt.legend(loc='upper left')
plt.show()
```

## Lift plot

Ploting the additional weight that load sensors would detect when the hive is lifted. The additional weight comes from the colony in the hive pushing on the load sensors during acceleration of the lift. The weight of the push will depend on how heavy the colony is, and how fast it is accelerated upwards. The additional weight is ploted as a function of the acceleration. The different lines represent different colony weights.

We use the equation for force F, mass m and acceleration a to find what the load sensors would meassure.

$F = ma$ (1)

First we use 1 to find the force that is distributed on the load cells. We use the mass of the colony: $ m_c $ or colonyMass in the code. We use the acceleration of the lift: $ a_t $ or accThief in the code.

$F_l = m_ca_t$

1 is used one more time to find the mass that the load cells would meassure at $F_l$. in this case the $a = g$ g is the acceleration from earths gravity. m is unknown.

$F_l = mg$

$m_ca_t = mg$

$m = \frac{m_ca_t}{g}$

We assume that they wil lift the hive from the ground and once the lift is finished they hold their arms in a streight down. we asume that that is around $\frac{1}{2}$ of their body at 0.9 m.

We assume that they lift at the bottom of the hive which stands on legs around 0.1 m above the ground. That means that initialy, to be carried away the hive will be lifted $0.9m - 0.1m = 0.8m$. We assume that the lift takes $1s$. If the speed of the lift was constant it would require a speed of $\frac{0.8m}{1s} = 0.8\frac{m}{s}$

We know that the speed will not be constant, and since we start from no speed, and end at no speed, we know that the actual speed must be higher than $0.8\frac{m}{s}$ at some point during the lift. Integrals could be used to get a more exact estimation of the actual top speed, but since this is just to get an estimation, we use $0.8\frac{m}{s}$ and calculate what acceleration you would need to achive that speed in half the lifting time: $0.25s$ This gives us an acceleration of: $\frac{0.8\frac{m}{s}}{0.25s} = 3.2\frac{m}{s^2}$
