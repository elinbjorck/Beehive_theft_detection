```python
import matplotlib.pyplot as plt
import numpy as np

time = [10, 20, 30, 40, 50, 60, 70, 80]
avg_current = [96.2, 76.7, 67.2, 63.6, 62.3, 60.7, 59.4, 57.7]

x = np.linspace(9.1, 80, 100)
def f(x):
    return 48 + 0.7 * ((5+3) / (x+5)) * 140
plt.scatter(time, avg_current)
plt.plot(x, f(x))
plt.xlabel('wait_time (s)')
plt.ylabel('average current (mA)')
plt.show()
```

```python
def f1(x, ib):
    return ib + 0.7 * ((5+3) / (x+5)) * (140 + (48 - ib))
plt.scatter(time, avg_current)
for ib in range(15, 50, 10):
 plt.plot(x, f1(x, ib), label='{} mA'.format(ib))
 plt.legend(loc='upper right')
plt.xlabel('wait_time (s)')
plt.ylabel('average current (mA)')
plt.show()
```

```python
def f2(x, icu):
    return 48 + 0.7 * ((5+3) / (x+5)) * icu
plt.scatter(time, avg_current)
for icu in range(80, 140, 10):
 plt.plot(x, f2(x, icu), label='{} mA'.format(icu))
 plt.legend(loc='upper right')
plt.xlabel('wait_time (s)')
plt.ylabel('average current (mA)')
plt.show()
```

```python
def f3(x, time_adv):
    return 48 + 0.7 * ((5 + time_adv) / (x+5)) * 140
plt.scatter(time, avg_current)
for time_adv in range(1, 4):
 plt.plot(x, f3(x, time_adv), label='{} mA'.format(time_adv))
 plt.legend(loc='upper right')
plt.xlabel('wait_time (s)')
plt.ylabel('average current (mA)')
plt.show()
```
